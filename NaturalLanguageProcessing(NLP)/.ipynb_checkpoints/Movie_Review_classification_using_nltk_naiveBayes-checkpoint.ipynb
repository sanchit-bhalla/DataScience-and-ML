### Clean data i.e remove stop words and punctuations and perform lemmatization also

```python
stop = set(stopwords.words('english'))
pun = string.punctuation
stop.update(pun)
stop
```

#### pos_tag gives parts of speech tag but in lemmatize we need different form of tag. So need to convert the tag given by pos_tag

```python
def get_simpler_pos(tag):
    if tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Assume rest everything is noun for simplicity
```

```python
###  eg
pos_tag(['better'])
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simpler_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
document = [(clean_review(doc), category) for doc, category in document]
```

```python
document[0]
```

### Separate train and test documents

```python
training_documents = document[0:1500]
testing_documents = document[1500:]
```

### Select features

```python
all_data = []
for doc in training_documents:
    all_data += doc[0]
```

```python
freq = nltk.FreqDist(all_data)
```

```python
freq.most_common(10)
```

```python
common = freq.most_common(3000)
```

```python
features = [i[0] for i in common]
```

```python
features
```

### Build Dataset in form of input required by the nltk NaiveBayes classifier

```python
def get_feature_dict(words):
    current_features = {}
    word_set = set(words)
    for f in features:
        current_features[f] = f in word_set
    return current_features
```

```python
training_data = [(get_feature_dict(doc),category) for doc, category in training_documents]
```

```python
testing_data = [(get_feature_dict(doc),category) for doc, category in testing_documents]
```

```python
training_data[0]
```

### Use naiveBayes Classifier

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```
